```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

tabel_data = pd.read_csv("../data/raw_data_machine_learning.csv")

tabel_data["Tanggal dibuat"] = pd.to_datetime(tabel_data["Tanggal dibuat"])
tabel_data["Tanggal diperbarui"] = pd.to_datetime(tabel_data["Tanggal diperbarui"])

tabel_data["Bahasa pemrograman"] = tabel_data["Bahasa pemrograman"].fillna("Tidak diketahui")

print("Total data siap visualisasi:")
print(tabel_data.shape)

plt.figure(figsize=(10, 6))
sns.countplot(y="Bahasa pemrograman", data=tabel_data, order=tabel_data["Bahasa pemrograman"].value_counts().index)
plt.title("Frekuensi bahasa pemrograman lengkap")
plt.xlabel("Jumlah repository")
plt.ylabel("Bahasa pemrograman")
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x="Jumlah bintang", y="Jumlah fork", data=tabel_data)
plt.title("Hubungan jumlah bintang dan fork")
plt.xlabel("Jumlah bintang")
plt.ylabel("Jumlah fork")
plt.show()
```
